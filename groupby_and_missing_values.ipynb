```python
# percentage of a country people use python
# first the value counts of countries
country_respondant=df['Country'].value_counts()
country_respondant
```

```python
country_use_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_use_python
```

```python
python_df=pd.concat([country_respondant,country_use_python],axis='columns',sort=False)
python_df
```

```python
python_df.rename(columns={'Country':'numRespondant','LanguageWorkedWith':'NumKnowsPython'},inplace=True)
```

```python
python_df
```

```python
# create new column 
python_df['pectnewpython']=(python_df['NumKnowsPython']/python_df['numRespondant'])*100
python_df
```

```python
# sort country by lagest percent who knew python
python_df.sort_values(by='pectnewpython',ascending=False,inplace=True)
python_df
```
